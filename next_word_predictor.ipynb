```python
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
tokenizer.word_index
```

```python
input_sequences=[]
for sentences in data.split("\n"):
  tokenized_sentences=(tokenizer.texts_to_sequences([sentences])[0])
  for i in range(1,len(tokenized_sentences)):
    input_sequences.append(tokenized_sentences[:i+1])
((input_sequences))
```

```python
max_len= max([len(x) for x in input_sequences])
max_len
```

```python
input_sequences=tensorflow.keras.utils.pad_sequences(input_sequences, maxlen=max_len, padding='pre')
```

```python
input_sequences
```

```python
X=input_sequences[:,:-1]

y=input_sequences[:,-1]
y
# this all is to covert in a supervisied learining task
# input we progressively add the words in each row and op is the new word
```

```python
X.shape
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=vocab_size)
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
model=Sequential()
model.add(Embedding(vocab_size, 100, input_length=max_len - 1))
model.add(LSTM(150)) # 150 nodes in each gate
model.add(Dense(vocab_size,activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.build(input_shape=(None, max_len - 1))
model.summary()
```

```python
model.fit(X, y, epochs=20)
```

```python
import numpy as np

text = "credit"
token_text = tokenizer.texts_to_sequences([text])[0]

#pad
padded_text= tensorflow.keras.utils.pad_sequences([token_text],maxlen=max_len, padding='pre')

pos=(np.argmax(model.predict(padded_text)))
for key ,val  in tokenizer.word_index.items():
  if val == pos:
    print(key)
```
